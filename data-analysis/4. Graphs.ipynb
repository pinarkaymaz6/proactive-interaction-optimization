```python
consecutive_days = list()

for chat_id, chat_df in messages.groupby('chat_id'):
    start = chat_df['date'].iloc[0]
    count=0
    for day, day_df in chat_df.groupby('just_date'):
        types = list(day_df['type'])
        if "message" in types or "callback_query" in types:
            count+= 1
        else:
            end=day
            delta = (end-start.date()).days
            break
    consecutive_days.append({"chat_id":chat_id, "consecutive_days":count})

consecutive_days = pd.DataFrame(consecutive_days)
#consecutive_days
```

```python
has_returned = list()

for chat_id, chat_df in messages.groupby('chat_id'):
    returned=True
    first = False
    for day, day_df in chat_df.groupby('just_date'):
        types = list(day_df['type'])
        if "message" in types or "callback_query" in types:
            if first:
                returned = True
                break
        else:
            first = True
            returned =False
    has_returned.append({"chat_id":chat_id, "has_returned":returned})

has_returned_df = pd.DataFrame(has_returned)
#has_returned_df
```

```python
active_hours=[]
for chat_id in users['chat_id']:
    chat_messages = messages[messages['chat_id']==chat_id]
    user_messages = chat_messages[(chat_messages["type"] == 'message') | (chat_messages["type"] == 'callback_query')]
    try:
        active_hour = user_messages['Hour'].value_counts().idxmax()
        active_hours.append({'chat_id':chat_id, 'active_hour':active_hour})
    except:
        active_hours.append({'chat_id':chat_id, 'active_hour':-1})

active_hours_df = pd.DataFrame(active_hours)
#active_hours_df
```

```python
active_days = list()
for chat_id, chat_df in messages.groupby('chat_id'):
    count = 0
    for day, day_df in chat_df.groupby('just_date'):
        user_input = day_df[(day_df["type"] == 'message') | (day_df["type"] == 'callback_query')]
        if len(user_input):
            count += 1
    active_days.append({"chat_id":chat_id, "active_days":count})
active_days_df = pd.DataFrame(active_days)
#active_days_df
```

```python
daily_usage = list()
for chat_id, chat_df in messages.groupby('chat_id'):
    total = 0
    for day, day_df in chat_df.groupby('just_date'):
        total += len(day_df)
    
    active_days= users['usage_time'].iloc[0]
    daily_usage.append({"chat_id":chat_id, "daily_usage":round(total/active_days,1)})

daily_usage = pd.DataFrame(daily_usage)
#daily_usage
```

```python
df_users_new=pd.merge(users, active_hours_df[["chat_id", "active_hour"]],on='chat_id', how='left')
df_users_new = pd.merge(df_users_new, daily_usage[["chat_id","daily_usage"]],on='chat_id', how='left')
df_users_new = pd.merge(df_users_new, active_days_df[["chat_id","active_days"]],on='chat_id', how='left')
df_users_new = pd.merge(df_users_new, has_returned_df[["chat_id","has_returned"]],on='chat_id', how='left')
df_users_new['interaction_rate'] =  round(df_users_new['active_days']/df_users_new['usage_time'],4)
df_users_new['interaction_rate'] = df_users_new['interaction_rate'].apply(lambda a: a if a<1 else 1.0 )
df_users_new = pd.merge(df_users_new, consecutive_days[["chat_id","consecutive_days"]],on='chat_id', how='left')
df_users_new['schedule_int'] = df_users_new['schedule'].apply(lambda a: int(a.split(':')[0]))
df_users_new
```

## Utterance Analysis


There are two types of user utterance format that Ada can process: Message and Callback objects. Callbacks are the inputs that user enters by selecting one of the buttons in Telegram's inline keyboard.Message objects are inputs that user entered in the typing bar. Message object can be either a Telegram command or a free text. Hence in overall, user can select a button (callbacks), send a free text or a Telegram command. The question is that what is the distribution of these input types. 

We should keep in mind that Ada's content is mood tracking and mood tracking functionality requires users to reply via callbacks. Hence, these statistics heavily depend on how the chatbot is designed. 

```python
callback_df = messages[messages['type']=='callback_query']
free_text_df = messages[(messages['type']=='message') & (messages['is_command']==False)]
command_df = messages[(messages['type']=='message') & (messages['is_command']==True)]
command_df['type']=command_df['type'].apply(lambda a: 'command')
utterence_df = pd.concat([callback_df, free_text_df, command_df], axis=0) 
#utterence_df
```

```python
plt.figure(figsize=(5,6))
sns.set(font_scale = 1.3)
sns.set_style("whitegrid")

utteranceplot= sns.countplot(x='type', data=utterence_df, palette='viridis')
utteranceplot.set_xticklabels(["Callback","Free Text","Command"])
total = len(utterence_df)
for p in utteranceplot.patches:
    height = p.get_height()
    width= p.get_width()
    x = p.get_x()
    utteranceplot.text(x+width/2., height + 10,f"{height} ({round((height/total)*100,1)}%)",ha="center", fontsize=12) 

plt.xlabel("Utterance Types")
plt.ylabel("Counts")
plt.title("Utterance Analysis", y=1.025, fontsize = 16)
plt.ylim(0,1700)
#plt.legend(loc=0)

utteranceplot_fig = utteranceplot.get_figure()
#utteranceplot_fig.savefig("../documentation/images/percentages.eps", format='eps', bbox_inches='tight')

```

```python
plt.figure(figsize=(12,6))
sns.set(font_scale = 1.3)
sns.set_style("whitegrid")

likert_colors = ['white', '#004466','#0077b3','#adebeb','#00cc99', '#008060']

dummy = pd.DataFrame([[1,0,0,4, 6], [0,0,1,5, 5], [0, 1, 0, 5, 5], [0,4,4,3,1],[0,0,2,3,6]],
                     columns=["strongly disagree", "disagree", "neutral", "agree", "strongly agree"],
                    index=["Instructions for participation was clear.", 
                           "I enjoyed tracking my mood daily.", 
                           "Timing of the daily reminders was convenient.",
                          "The frequency of rating prompts was  too much.",
                          "I'm satisfied with the overall performance."])


middles = dummy[["strongly disagree", "disagree"]].sum(axis=1)+dummy["neutral"]*.5
longest = middles.max()+1
complete_longest = dummy.sum(axis=1).max()+1
dummy.insert(0, '', (middles - longest).abs())

dummy.plot.barh(stacked=True, color=likert_colors, edgecolor='none', legend=False)
z = plt.axvline(longest, linestyle='-', color='black', alpha=.8)
z.set_zorder(-1)

plt.xlim(0, complete_longest+5)
xvalues = [int(i) for i in range(0,complete_longest+6,1)]
xlabels = [str(int(x-longest)) for x in xvalues]
plt.xticks(xvalues, xlabels)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=11)

plt.title('"How would you describe your experience with chatting with Ada?"', y=1.025, fontsize = 16)

#plt.savefig("../documentation/images/likert.png", bbox_inches='tight')
#plt.show()
```
